```python
import os
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# opening csv file provided by you
csv_file = 'C:\\Users\\Teqs\\Downloads\\nyc_311_data_2022.csv'
df = pd.read_csv("nyc_311_data_2022.csv")
df
```

```python
# task 1 b.
# Filter the 'created_date' column to keep the data just for December 2022

only_dec  = '2022-12'
df = df[df['created_date'].str[:7] == only_dec]
df
```

```python
# task 1 a.
# Remove the data records where either the 'incident_zip', 'latitude', or 'longitude' columns have missing values or NaN

df_cleaned = df.dropna(subset=['incident_zip'])
df_cleaned
```

```python
# task 2
# Get the total number of requests for each complaint type ('complaint_type'), and plot a pie
# chart depicting the % share of each top five complaint type vs the rest within the total activity.

complaint_counts = df['complaint_type'].value_counts()
top_five_complaints = complaint_counts.nlargest(5)
rest_complaints_count = complaint_counts[~complaint_counts.index.isin(top_five_complaints.index)].sum()
combined_counts = pd.concat([top_five_complaints, pd.Series({'Other': rest_complaints_count})])
plt.figure(figsize=(7, 7))
plt.pie(combined_counts, labels=combined_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Top Five Complaint Types vs. Rest')
plt.show()
```

```python
# task 3
# Plot the hourly distribution of total complaints activity in a bar plot. 
# (Use the 'created_date' column to get the hour for each complaint record)

df['created_date'] = pd.to_datetime(df['created_date'])
df.loc[:, 'hour'] = df['created_date'].dt.hour
hourly_complaints = df.groupby('hour').size()
plt.figure(figsize=(9, 5))
hourly_complaints.plot(kind='bar', color='green')
plt.xlabel('Hour of Day')
plt.ylabel('Total Complaints Activity')
plt.title('Hourly Distribution of Total Complaints Activity')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# task 3a
# Plot the relative proportions (%) of the top three complaint types for each hour, where the
# percentages are calculated based on the total count of these three types only. Each of the
# three complaint types can be visualized by a separate color. (Use the 'created_date'
# column to get the hour for each complaint record).

df['created_date'] = pd.to_datetime(df['created_date'])
df['hour'] = df['created_date'].dt.hour
hourly_top_complaints = df.groupby(['hour', 'complaint_type']).size().unstack()
top_three_types = df['complaint_type'].value_counts().nlargest(3).index
hourly_top_complaints = hourly_top_complaints[top_three_types]
hourly_proportions = hourly_top_complaints.apply(lambda x: x / x.sum() * 100, axis=1)
plt.figure(figsize=(9, 5))
hourly_proportions.plot(kind='bar', stacked=True)
plt.xlabel('Hour of Day')
plt.ylabel('Relative Proportions (%)')
plt.title('Relative Proportions of Top Three Complaint Types for Each Hour')
plt.xticks(rotation=0)
plt.legend(title='Complaint Type')
plt.tight_layout()
plt.show()
```

```python
# task 4a
# Get the processing time for each data record. Columns 'created_date' represents the
# time when the complaint was made and 'closed_date' when it was resolved. 

df['created_date'] = pd.to_datetime(df['created_date'])
df['closed_date'] = pd.to_datetime(df['closed_date'])
df['processing_time'] = df['closed_date'] - df['created_date']
df[['created_date', 'closed_date', 'processing_time']]
```

```python
# task 4b
# Get the median processing time by each city agency and sort to get the three fastest and
# three slowest city agencies (the 'agency_name' column represents the city agency)

median_processing_time = df.groupby('agency_name')['processing_time'].median()
sorted_agencies = median_processing_time.sort_values()
fastest_agencies = sorted_agencies.head(3)
slowest_agencies = sorted_agencies.tail(3)
print("Three Fastest City Agencies:")
print(fastest_agencies)
print("\nThree Slowest City Agencies:")
print(slowest_agencies)
```
